## OpenCV를 활용한 이미지 크기 및 픽셀 정보 확인

```python
import cv2
image = cv2.imread('cat.jpg')

# 이미지 픽셀 수 확인
print(image.shape)
# 이미지 크기 확인
print(image.size)

# 이미지 Numpy 객체의 특정 픽셀을 가리키는 법
px = image[100, 100]

# B, G, R 순서로 출력. (단, GrayScale인 경우에는 B, G, R 로 구분되지 않음.)
print(px)

# R 값만 출력하기
print(px[2])
```

## OpenCV를 활용한 특정 범위 픽셀 변경

```python
import cv2
import matplotlib.pyplot as plt
import time

image = cv2.imread('cat.jpg')

start_time = time.time()
for i in range(0, 100):
  for j in range(0, 100):
    image[i, j] = [255, 255, 255]
print("--- %s seconds ---" % (time.time() - start_time))

# Slicing 연산을 사용하면 연산 처리 속도가 훨씬 빠르다
start_time = time.time()
image[0:100, 0:100] = [255, 255, 255]
print("--- %s seconds ---" % (time.time() - start_time))

plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```

## OpenCV를 활용한 ROI(Region of Interest : 관심 있는 영역) 추출

```python
import cv2
import matplotlib.pyplot as plt

image = cv2.imread('cat.jpg')

# Numpy slicing : ROI 처리 가능
roi = image[300:400, 150:250]

# ROI 단위로 이미지 복사하기
image[0:100, 0:100] = roi

plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```

## OpenCV를 활용한 픽셀별 색상 다루기

```python
import cv2
import matplotlib.pyplot as plt

image = cv2.imread('cat.jpg')
# B, G, R 에서 R값을 0으로 만들면 이미지에서 붉은 색이 빠짐.
image[:, :, 2] = 0

plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```
